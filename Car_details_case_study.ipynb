```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sc
```

```python
# to read the csv file
df = pd.read_csv("Car_details.csv")
df
```

```python
# To get the first 10 readings
df.head(10)
```

```python
# To get the datatypes of variables
df.info()
```

```python
# just using isna() will give True False Values
df.isna()
```

```python
# Will give the sum of all numerice values
# isnull ans isna are both same function which gives us the count of null values in each column
df.isna().sum()
```

```python
# As we have mileage column in that the enteries are like 25.50kmpl
# So this will just give us the the numeric value and delete the string value
df["mileage"]=df["mileage"].str[0:4]
df["mileage"]
```

```python
# Will give us the count of each unique value in mileage column
df.mileage.value_counts()
```

```python
# To change the datatype of mileage from object to float
df.mileage=df.mileage.astype("float")
```

```python
# Mileage datatype is changes now
df.info()
```

```python
# heat map for all columns in dataset
# isnull gives null values 
# we can see the visualization of null values
sns.heatmap(df.isnull())
```

```python
# As we have engine column in that the enteries are like 1284CC
# So this will just give us the the numeric value and delete the string value
df["engine"]=df["engine"].str[0:4]
df["engine"]
```

```python
# Will give us the count of each unique value in engine column
df.engine.value_counts()
```

```python
# To change the datatype of engine from object to float
df.engine=df.engine.astype("float")
```

```python
# Engine datatype is changes now
df.info()
```

```python
# As we have max_power column in that the enteries are like 128bhp
# So this will just give us the the numeric value and delete the string value
# This is another way to get only numeric values
df["max_power"]=pd.to_numeric(df["max_power"].str.strip("bhp "))
df["max_power"]
```

```python
# Will give us the count of each unique value in max_power column
df.max_power.value_counts()
```

```python
# To change the datatype of max_power from object to float
df.max_power=df.max_power.astype("float")
```

```python
# max_power datatype is changes now
df.info()
```

```python
# Will give the sum of all numerice values
# isnull gives us count of null values
df.isnull().sum()
```

```python
# Will replace the null values with given values for engine column
df["engine"]=df["engine"].fillna(df["engine"].mean())
# Will replace the null values with given values for max_power column
df["max_power"]=df["max_power"].fillna(df["max_power"].mean())
# Will replace the null values with given values for seats column
df["seats"]=df["seats"].fillna(df["seats"].mean())
# Will replace the null values with given values for mileage column
df["mileage"]=df["mileage"].fillna(df["mileage"].mean())
```

```python
# Will give the sum of all numerice values
# isnull gives us count of null values
df.isnull().sum()
```

```python
# heat map for all columns in dataset
# isnull gives null values 
# we can see the visualization of null values
# now all null values are cleared
sns.heatmap(df.isnull())
```

```python
# To get the column names
df.columns
```

```python
# Barplot between owners and selling price 
# the line on each bar showa the fuel used for the car
sns.barplot(data=df,x="owner",y="selling_price")
plt.xticks(rotation=90)    # will rotate the names on x-axis
plt.title("Fuel Vs Selling Price")
plt.show()
```

```python
# seperating the numerical and categorical data columns in dataset
# checking for the categorical data
cat_col=[]
for i in df.dtypes.index:
    if df.dtypes[i]=='object':
        cat_col.append(i)
print("Categorical columns present in the dataset are : ",cat_col)
# checking for the numerical data
num_col=[]
for i in df.dtypes.index:
    if df.dtypes[i]!='object':
        num_col.append(i)
print("Numerical columns present in the dataset are : ",num_col)
```

```python
# distribution od skewness for all columns which are numeric in which direction does the data go
plt.figure(figsize=(12,12),facecolor="white")
plt.suptitle("Visualize the distribution os skewnwss in numerical variable",fontsize=20)
plotnumber=1
for column in num_col:
    if plotnumber<=6:
        zx=plt.subplot(3,2,plotnumber)
        sns.distplot(df[column],color="teal")
        plt.xlabel(column,fontsize=15)
    plotnumber+=1
plt.tight_layout()
        
```

```python
# distribution od skewness for all columns which are numeric in which direction does the data go
# using boxplot to see the outliers
plt.figure(figsize=(12,12),facecolor="white")
plt.suptitle("Visualize the outliers in numerical variable using boxplot",fontsize=20)
plotnumber=1
for column in num_col:
    if plotnumber<=6:
        zx=plt.subplot(3,2,plotnumber)
        sns.boxplot(df[column],color="teal")
        plt.xlabel(column,fontsize=15)
    plotnumber+=1
plt.tight_layout()
```
